```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
x=pd.read_csv("E:\\chrome_downloads\\Microsoft Dataset.csv")
x.head()
```

```python
x1=x[["High","Low"]]
y1=x["Open"]
x1=x1.to_numpy()
y1=y1.to_numpy()
y1=y1.reshape(-1,1)
x1_train,x1_test,y1_train,y1_test=train_test_split(x1,y1,test_size=0.30,random_state=42)
model=LinearRegression()
model.fit(x1_train,y1_train)
y1pred=model.predict(x1_test)
data=pd.DataFrame(data={"predicted rate":y1pred.flatten()})
print(data.head())
```
